# **CAPSTONE ASSIGNMENT**

```python
# Colab-ready: Corrected Capstone: Multi-Strategy Logistics Optimization
# - Uses all-pairs shortest paths (Dijkstra from each node)
# - Time-window feasibility uses travel along shortest routes between visited nodes
# - Capacity enforced in subset search
# - Visualization: network plot + priority bar + route-cost plot + best-route overlay

# Install (only if missing)
# !pip install networkx

import heapq
import itertools
import time
import math
import matplotlib.pyplot as plt

# -------------------------
# Problem inputs
# -------------------------
nodes = ["Warehouse", "A", "B", "C", "D", "E"]

# Directed weighted graph (missing edges = no direct road)
graph = {
    "Warehouse": {"A": 4, "B": 6, "C": 8},
    "A": {"B": 2, "C": 3, "D": 7},
    "B": {"C": 2, "D": 5, "E": 8},
    "C": {"D": 2, "E": 6},
    "D": {"E": 3},
    "E": {}
}

# deliveries: (location, priority (1 high), earliest_time, latest_time, demand)
deliveries = [
    ("A", 2, 9, 12, 1),
    ("B", 3, 10, 15, 1),
    ("C", 1, 8, 11, 1),
    ("D", 4, 13, 17, 2),
    ("E", 2, 14, 18, 1)
]

VEHICLE_CAPACITY = 3
START_TIME = 8
SERVICE_TIME = 0.5  # time spent at a delivery

# -------------------------
# Helper: Dijkstra (single-source). Returns dist dict and parent for path reconstruction
# -------------------------
def dijkstra_with_parent(graph, start):
    dist = {node: math.inf for node in graph}
    parent = {node: None for node in graph}
    dist[start] = 0
    pq = [(0, start)]
    while pq:
        d, u = heapq.heappop(pq)
        if d > dist[u]:
            continue
        for v, w in graph.get(u, {}).items():
            nd = d + w
            if nd < dist[v]:
                dist[v] = nd
                parent[v] = u
                heapq.heappush(pq, (nd, v))
    return dist, parent

# Precompute all-pairs shortest distances and parent maps (for path reconstruction)
all_pairs_dist = {}
all_pairs_parent = {}
for n in nodes:
    dist, parent = dijkstra_with_parent(graph, n)
    all_pairs_dist[n] = dist
    all_pairs_parent[n] = parent

# Function to reconstruct shortest path from u to v using parent maps from u
def reconstruct_path(u, v):
    # uses parent map computed from u
    parent = all_pairs_parent[u]
    if parent[v] is None and v != u:
        # maybe there is no path (or v==u)
        if u == v:
            return [u]
        return None
    # follow parents backwards
    path = []
    cur = v
    while cur is not None:
        path.append(cur)
        if cur == u:
            break
        cur = parent[cur]
    if path[-1] != u:
        # no path
        return None
    return list(reversed(path))

# -------------------------
# Strategy 1: Greedy by (priority asc, latest_time asc)
# -------------------------
def greedy_select(deliveries, capacity):
    sorted_d = sorted(deliveries, key=lambda x: (x[1], x[3]))  # priority asc (1 best), then earliest deadline
    chosen = sorted_d[:capacity]
    return chosen

greedy_choice = greedy_select(deliveries, VEHICLE_CAPACITY)
print("Greedy selection (priority-first):", greedy_choice)

# -------------------------
# Time-window feasibility for a sequence (uses shortest-path travel between successive visited nodes)
# -------------------------
def windows_feasible(start_time, sequence):
    # sequence: list of delivery tuples in execution order
    time = start_time
    prev = "Warehouse"
    for loc, priority, earliest, latest, demand in sequence:
        travel = all_pairs_dist[prev].get(loc, math.inf)
        if math.isinf(travel):
            return False  # unreachable
        arrival = time + travel
        if arrival > latest:
            return False
        time = max(arrival, earliest) + SERVICE_TIME
        prev = loc
    # return to Warehouse (optional check) - we don't require return time window
    return True

print("Window feasible for greedy choice?", windows_feasible(START_TIME, greedy_choice))

# -------------------------
# Routes: generate permutations for a given list of locations and evaluate using shortest-path distances
# -------------------------
def generate_routes(loc_list):
    perms = list(itertools.permutations(loc_list))
    routes = []
    for p in perms:
        # route = sequence of nodes including warehouse start and warehouse end
        route = ["Warehouse"] + list(p) + ["Warehouse"]
        routes.append(route)
    return routes

def route_total_cost_shortest(route):
    # Sum shortest-path distances between consecutive nodes; if any leg unreachable -> inf
    total = 0
    for i in range(len(route)-1):
        u, v = route[i], route[i+1]
        d = all_pairs_dist[u].get(v, math.inf)
        if math.isinf(d):
            return math.inf
        total += d
    return total

# evaluate permutations for greedy chosen set
chosen_locs = [d[0] for d in greedy_choice]
routes = generate_routes(chosen_locs)
print(f"Generated {len(routes)} permutations for chosen deliveries.")

# compute costs
route_costs = [route_total_cost_shortest(r) for r in routes]
finite_costs = [c for c in route_costs if not math.isinf(c)]
if finite_costs:
    best_idx = route_costs.index(min(finite_costs))
    best_route = routes[best_idx]
    best_cost = route_costs[best_idx]
else:
    best_route, best_cost = None, math.inf

print("Best route (based on shortest-path distances) for greedy set:", best_route, " cost:", best_cost)

# -------------------------
# Brute-force: evaluate all subsets up to capacity and all permutations, also check time-window feasibility
# -------------------------
def all_subsets_best_route(deliveries, capacity, start_time):
    best = None
    bestc = math.inf
    n = len(deliveries)
    for mask in range(1, 1<<n):
        subset = []
        total_demand = 0
        for i in range(n):
            if (mask>>i) & 1:
                subset.append(deliveries[i])
                total_demand += deliveries[i][4]
        if total_demand > capacity:
            continue
        locs = [d[0] for d in subset]
        for perm in itertools.permutations(subset):
            # perm is a sequence of delivery tuples in order
            # check time windows feasibility (starts at warehouse)
            if not windows_feasible(start_time, list(perm)):
                continue
            route = ["Warehouse"] + [d[0] for d in perm] + ["Warehouse"]
            cost = route_total_cost_shortest(route)
            if cost < bestc:
                bestc = cost
                best = (route, list(perm), cost)
    return best

t0 = time.time()
bf = all_subsets_best_route(deliveries, VEHICLE_CAPACITY, START_TIME)
t1 = time.time()
print("Brute-force best (subset+permutations) result:", bf)
print("Brute-force time: {:.4f} s".format(t1 - t0))

# -------------------------
# Visualization
# - Use fixed coordinates for nodes for a nice-looking plot (you can change positions)
# - Draw directed edges (as lines with arrowheads approximated)
# - Overlay best route in red
# -------------------------
positions = {
    "Warehouse": (0, 0),
    "A": (2, 4),
    "B": (5, 4),
    "C": (7, 1),
    "D": (4, -2),
    "E": (1, -1)
}

# Helper to draw arrow (line with arrowhead)
def draw_edge(ax, u, v, weight, color='gray', alpha=0.8):
    x1, y1 = positions[u]
    x2, y2 = positions[v]
    # shorten slightly so arrowheads don't overlap node markers
    dx = x2 - x1
    dy = y2 - y1
    dist = math.hypot(dx, dy)
    if dist == 0:
        return
    shrink = 0.12
    startx = x1 + shrink * dx / dist
    starty = y1 + shrink * dy / dist
    endx = x2 - shrink * dx / dist
    endy = y2 - shrink * dy / dist
    ax.annotate("",
                xy=(endx, endy), xytext=(startx, starty),
                arrowprops=dict(arrowstyle="->", color=color, lw=1, alpha=alpha),
                )
    # mid label for weight
    mx = (startx + endx) / 2
    my = (starty + endy) / 2
    ax.text(mx, my, f"{weight}", fontsize=8, color='black', ha='center', va='center', bbox=dict(facecolor='white', alpha=0.6, boxstyle='round,pad=0.1'))

# Create figure with subplots: graph left, priority bar top-right, route cost bottom-right
fig = plt.figure(figsize=(14,6))
ax_graph = fig.add_axes([0.03, 0.1, 0.58, 0.85])   # left large area
ax_bar = fig.add_axes([0.64, 0.55, 0.33, 0.4])     # top-right
ax_cost = fig.add_axes([0.64, 0.08, 0.33, 0.4])    # bottom-right

# Draw nodes
for n, (x,y) in positions.items():
    ax_graph.scatter(x, y, s=300, zorder=5, color='lightblue', edgecolor='k')
    ax_graph.text(x, y, n, ha='center', va='center', fontsize=10, weight='bold')

# Draw directed edges with weights (from graph)
for u, edges in graph.items():
    for v, w in edges.items():
        draw_edge(ax_graph, u, v, w, color='gray')

ax_graph.set_title("Road Network (nodes & directed edges with distances)")
ax_graph.axis('off')

# Overlay best route if exists (use red thick lines over shortest path edges)
def overlay_route(ax, route, color='red'):
    if not route:
        return
    # construct full path by concatenating shortest-path legs
    full_path = []
    for i in range(len(route)-1):
        u, v = route[i], route[i+1]
        sp = reconstruct_path(u, v)
        if sp is None:
            full_path = None
            break
        # append (for consecutive legs avoid duplication)
        if not full_path:
            full_path.extend(sp)
        else:
            # avoid repeating start node
            full_path.extend(sp[1:])
    if not full_path:
        return
    # draw the polyline segments
    for i in range(len(full_path)-1):
        x1, y1 = positions[full_path[i]]
        x2, y2 = positions[full_path[i+1]]
        ax.plot([x1, x2], [y1, y2], color=color, linewidth=3, zorder=6, alpha=0.9)
        ax.scatter([x1, x2], [y1, y2], s=320, facecolors='none', edgecolors=color, linewidths=1.5, zorder=7)

# choose route to overlay: brute-force best if available else greedy best
overlay_choice_route = None
if bf is not None:
    overlay_choice_route = bf[0]
elif best_route:
    overlay_choice_route = best_route

overlay_route(ax_graph, overlay_choice_route, color='red')

# Priority bar chart
locs = [d[0] for d in deliveries]
pr = [d[1] for d in deliveries]
ax_bar.bar(locs, pr, color='skyblue', edgecolor='k')
ax_bar.set_title("Delivery Priorities (1 = highest)")
ax_bar.set_xlabel("Location")
ax_bar.set_ylabel("Priority (lower = higher)")
ax_bar.invert_yaxis()  # so 1 appears at top visually (optional)

# Route costs for permutations (chosen set)
ax_cost.plot(range(len(route_costs)), route_costs, marker='o', linestyle='-', color='tab:blue')
ax_cost.set_title("Route costs for permutations (chosen greedy deliveries)")
ax_cost.set_xlabel("Permutation index")
ax_cost.set_ylabel("Cost (shortest-path distance)")
ax_cost.grid(True)

plt.suptitle("Capstone: Multi-Strategy Logistics Optimization — Visualization", fontsize=14)
plt.show()

# -------------------------
# Complexity summary and final prints
# -------------------------
print("\n--- Results Summary ---")
print("Greedy selected deliveries:", greedy_choice)
if best_route:
    print("Best route for greedy set (shortest-path based):", best_route, "cost:", best_cost)
if bf:
    print("Global best (subset + permutations) w/ time windows:", bf)
print("All-pairs shortest-path distances (sample):")
for u in all_pairs_dist:
    print(f" From {u}: {all_pairs_dist[u]}")

print("\n--- Complexity Summary ---")
print("Greedy selection: O(n log n) (sorting).")
print("All-pairs shortest paths (our precompute): O(V * (E + V log V)) using repeated Dijkstra.)")
print("Permutations for k chosen deliveries: O(k!) to evaluate.")
print("Brute-force subsets + permutations: O(2^n * n!) worst-case (infeasible for moderate n).")
print("This problem contains TSP/VRP-like subproblems → combinatorial/NP-hard behavior.")

```

```python
# -----------------------------------------------------------

# -----------------------------------------------------------

import itertools
import math
import time
import networkx as nx
import matplotlib.pyplot as plt

# ---------------------------------
# MAKE GRAPH UNDIRECTED FOR VALID ROUTES
# ---------------------------------
graph = {
    "Warehouse": {"A": 4, "B": 6, "C": 8},
    "A": {"B": 2, "C": 3, "D": 7},
    "B": {"C": 2, "D": 5, "E": 8},
    "C": {"D": 2, "E": 6},
    "D": {"E": 3},
    "E": {}
}

# Convert to undirected
G = nx.Graph()
for u in graph:
    for v, w in graph[u].items():
        G.add_edge(u, v, weight=w)

nodes = list(G.nodes())

deliveries = [
    ("A", 2, 9, 12, 1),
    ("B", 3, 10, 15, 1),
    ("C", 1, 8, 11, 1),
    ("D", 4, 13, 17, 2),
    ("E", 2, 14, 18, 1),
]

VEHICLE_CAPACITY = 3
START_TIME = 8

# shortest paths
all_pairs = dict(nx.all_pairs_dijkstra_path_length(G))
all_paths = dict(nx.all_pairs_dijkstra_path(G))

# greedy
def greedy(deliveries):
    return sorted(deliveries, key=lambda x: (x[1], x[3]))[:VEHICLE_CAPACITY]

greedy_choice = greedy(deliveries)
print("Greedy:", greedy_choice)

chosen_locs = [d[0] for d in greedy_choice]

# generate valid permutations
routes = []
costs = []

for perm in itertools.permutations(chosen_locs):
    route = ["Warehouse"] + list(perm) + ["Warehouse"]
    valid = True
    total = 0
    for i in range(len(route)-1):
        u = route[i]
        v = route[i+1]
        if v not in all_pairs[u]:
            valid = False
            break
        total += all_pairs[u][v]
    if valid:
        routes.append(route)
        costs.append(total)

print("Valid routes:", len(routes))

best_route = routes[costs.index(min(costs))]
best_cost = min(costs)
print("Best route:", best_route, "cost =", best_cost)

# ------------------ VISUALIZATION ------------------

plt.figure(figsize=(15,6))
gs = plt.GridSpec(2, 3)

# Graph
ax1 = plt.subplot(gs[:, 0])
pos = nx.spring_layout(G, seed=10)
nx.draw(G, pos, with_labels=True, node_size=1200, font_size=10,
        node_color="lightblue", ax=ax1)

labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels, ax=ax1)

# highlight best route
highlight_edges = []
for i in range(len(best_route)-1):
    u = best_route[i]
    v = best_route[i+1]
    sub = all_paths[u][v]
    for j in range(len(sub)-1):
        highlight_edges.append((sub[j], sub[j+1]))

nx.draw_networkx_edges(G, pos, edgelist=highlight_edges, edge_color="red", width=3, ax=ax1)

# priority bar
ax2 = plt.subplot(gs[0,1])
locs = [d[0] for d in deliveries]
pr = [d[1] for d in deliveries]
ax2.bar(locs, pr, color="skyblue", edgecolor="black")
ax2.invert_yaxis()
ax2.set_title("Priorities")

# permutation costs
ax3 = plt.subplot(gs[1,1])
ax3.plot(range(len(costs)), costs, marker='o', color='green')
ax3.set_title("Permutation Costs")
ax3.grid(True)

# summary block
ax4 = plt.subplot(gs[:,2])
ax4.axis("off")
ax4.text(0, 0.5,
         f"Best Route: {best_route}\nCost: {best_cost}\n\nGreedy: {greedy_choice}\nValid perms: {len(routes)}",
         fontsize=11)
plt.tight_layout()
plt.show()
```
